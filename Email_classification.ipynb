```python
#splitting dataset into x and y
x=data[data.columns[:-1]]
y=data[data.columns[-1]]
```

```python
x
```

Scaling dataframe X

```python
#Scaling dataframe X since its values in column 55 and 56 are higher compared to other columns
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
(scaler.fit(x))
X=scaler.transform(x)
```

```python
X
```

Implimenting perceptron algoritham

```python
#Implimenting perceptron algoritham
from sklearn.linear_model import Perceptron
model=Perceptron()
model.fit(X,y)
```

```python
#defining model evaluation method and printing model score
from numpy import mean
from numpy import std
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
evl = RepeatedStratifiedKFold(n_splits=10, n_repeats=3)
score = cross_val_score(model, X, y, scoring='accuracy', cv=evl)
mean(score)
```

Tuning hyperparamenters epochs and learning rate

```python
#Tuning hyperparamenters epochs and learning rate
from sklearn.model_selection import GridSearchCV
grid = dict()
grid['max_iter'] = [1, 10, 100, 1000, 10000]
# define search
search = GridSearchCV(model, grid, scoring='accuracy', cv=evl)
# perform the search
results = search.fit(X, y)
```

```python
results
```

```python
print(results.best_score_)
print(results.best_params_)
```

```python
grid['eta0'] = [0.0001, 0.001, 0.01, 0.1, 1.0]
search = GridSearchCV(model, grid, scoring='accuracy', cv=evl, n_jobs=-1)
results = search.fit(X, y)
print(results.best_score_)
print(results.best_params_)
```
